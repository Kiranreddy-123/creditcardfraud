```python
#Importing the Dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset
credit_card_data=pd.read_csv("creditcard.csv")
```

```python
#first five rows
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#dataset information
credit_card_data.info()
```

```python
#checking n0. of missing values
credit_card_data.isnull().sum()
```

```python
#distribution of transactions
credit_card_data['Class'].value_counts()
```

```python
#dataset is unbalanced
# 0-->Normal transaction
# 1-->Fradulent transaction
#seperate the data for analysis

legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#stastical methods of data

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare values of both transaction
credit_card_data.groupby('Class').mean()
```

```python
# Under sampling
legit_sample=legit.sample(n=492)
```

```python
#concatenate two dataframes

new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#splitting the data into Features and targets
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#slit the data into train & test data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#model training
#Logistic Regression

model=LogisticRegression()
```

```python
#training the model(logstic reg)

model.fit(X_train,Y_train)
```

```python
#model evalution
#Accuracy score

X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training data:',training_data_accuracy)
```

```python
#Accuracy on test data

X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy score of test data:',test_data_accuracy)
```
